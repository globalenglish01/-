````python
# #自动往chatgpt里发送.md文件里的指令

# # pip install playwright
# #pip install playwright pyperclip

# # python -m playwright install chromium
# # python -m playwright --version


# # from playwright.sync_api import sync_playwright

# # with sync_playwright() as p:
# #     browser = p.chromium.launch()
# #     print("Playwright OK")
# #     browser.close()



# # from playwright.sync_api import sync_playwright
# # import time
# # import random
# # import os
# # import pyperclip

# # CHATGPT_URL = "https://chat.openai.com"
# # AUTH_FILE = "auth.json"

# # prompt = "解释什么是 AWS Glue"


# # # -------------------------
# # # 等待输入框
# # # -------------------------
# # def wait_for_input(page):

# #     print("等待输入框...")

# #     page.wait_for_selector(
# #         'div[contenteditable="true"]',
# #         timeout=120000
# #     )

# #     print("输入框已就绪")


# # # -------------------------
# # # 发送 prompt
# # # -------------------------
# # def send_prompt(page, text):

# #     input_box = page.locator('div[contenteditable="true"]').first

# #     input_box.wait_for(state="visible", timeout=60000)

# #     input_box.click()

# #     input_box.fill(text)

# #     send_button = page.locator('button[data-testid="send-button"]')

# #     send_button.wait_for(state="visible")

# #     send_button.click()


# # # -------------------------
# # # 获取 AI 回复
# # # -------------------------
# # def get_last_answer(page):

# #     messages = page.locator(
# #         'div[data-message-author-role="assistant"]'
# #     )

# #     count = messages.count()

# #     if count == 0:
# #         return ""

# #     return messages.nth(count - 1).inner_text()


# # # -------------------------
# # # 检测 AI 输出结束
# # # -------------------------
# # def wait_for_answer_finish(page, timeout=180):

# #     last_text = ""

# #     start = time.time()

# #     while True:

# #         text = get_last_answer(page)

# #         if text == "":
# #             time.sleep(2)
# #             continue

# #         if text == last_text:
# #             return text

# #         last_text = text

# #         if time.time() - start > timeout:
# #             return text

# #         time.sleep(3)


# # # -------------------------
# # # 保存登录状态
# # # -------------------------
# # def save_login(context):

# #     print("请手动登录 ChatGPT")
# #     input("登录完成后按回车保存 session")

# #     context.storage_state(path=AUTH_FILE)

# #     print("登录状态已保存")


# # import re

# # def extract_prompts(md_file):

# #     with open(md_file, "r", encoding="utf-8") as f:
# #         content = f.read()

# #     pattern = r"```(.*?)```"

# #     prompts = re.findall(pattern, content, re.S)

# #     return [p.strip() for p in prompts if p.strip()]

# # # -------------------------
# # # 主程序
# # # -------------------------
# # def main():
# #     # prompts = extract_prompts("lesson.md")

# #     # print("发现 prompt 数量:", len(prompts))

# #     # for i, prompt in enumerate(prompts):

# #     #     print("\n====================")
# #     #     print(f"Prompt {i+1}")
# #     #     print("====================")

# #     #     send_prompt(page, prompt)

# #     #     print("等待 AI 回复...")

# #     #     answer = wait_for_answer_finish(page)

# #     #     print(answer[:500])

# #     #     with open("result.txt", "a", encoding="utf-8") as f:

# #     #         f.write("\n\n====================\n")
# #     #         f.write(f"PROMPT {i+1}\n\n")
# #     #         f.write(answer)

# #     #     time.sleep(8)


# #     with sync_playwright() as p:

# #         browser = p.chromium.launch(
# #             headless=False,
# #             args=["--disable-blink-features=AutomationControlled"]
# #         )

# #         # 读取登录状态
# #         if os.path.exists(AUTH_FILE):

# #             context = browser.new_context(
# #                 storage_state=AUTH_FILE
# #             )

# #         else:

# #             context = browser.new_context()

# #         page = context.new_page()

# #         # 隐藏 webdriver
# #         page.add_init_script("""
# #         Object.defineProperty(navigator, 'webdriver', {
# #             get: () => undefined
# #         })
# #         """)

# #         print("打开 ChatGPT...")

# #         page.goto(CHATGPT_URL)

# #         wait_for_input(page)

# #         # 首次运行保存登录
# #         if not os.path.exists(AUTH_FILE):

# #             save_login(context)

# #         global prompt

# #         for i in range(5):

# #             print("\n===================")
# #             print(f"第 {i+1} 轮")
# #             print("===================")

# #             print("发送:", prompt)

# #             send_prompt(page, prompt)

# #             print("等待 AI 生成...")

# #             answer = wait_for_answer_finish(page)

# #             if answer == "":

# #                 print("未检测到回答")

# #                 continue

# #             print("\nAI回答前500字:\n")

# #             print(answer[:500])

# #             pyperclip.copy(answer)

# #             print("\n回答已复制到剪贴板")

# #             prompt = f"继续详细解释：\n{answer[:200]}"

# #             sleep_time = random.randint(8, 15)

# #             print(f"\n等待 {sleep_time} 秒...")

# #             time.sleep(sleep_time)

# #         print("\n任务结束")

# #         browser.close()


# # if __name__ == "__main__":

# #     main()

# from playwright.sync_api import sync_playwright
# import time
# import random
# import os
# import re

# CHATGPT_URL = "https://chat.openai.com"
# AUTH_FILE = "auth.json"
# RESULT_FILE = "result.txt"
# PROGRESS_FILE = "progress.txt"


# # ----------------------------
# # 提取 markdown prompt
# # ----------------------------
# def extract_prompts(md_file):

#     with open(md_file, "r", encoding="utf-8") as f:
#         content = f.read()

#     pattern = r"```(.*?)```"

#     prompts = re.findall(pattern, content, re.S)

#     return [p.strip() for p in prompts if p.strip()]


# # ----------------------------
# # 读取进度
# # ----------------------------
# def load_progress():

#     if os.path.exists(PROGRESS_FILE):

#         with open(PROGRESS_FILE) as f:
#             return int(f.read().strip())

#     return 0


# # ----------------------------
# # 保存进度
# # ----------------------------
# def save_progress(i):

#     with open(PROGRESS_FILE, "w") as f:
#         f.write(str(i))


# # ----------------------------
# # 等待输入框
# # ----------------------------
# def wait_for_input(page):

#     page.wait_for_selector(
#         'div[contenteditable="true"]',
#         timeout=120000
#     )


# # ----------------------------
# # 发送 prompt
# # ----------------------------
# # def send_prompt(page, text):

# #     input_box = page.locator('div[contenteditable="true"]').first

# #     input_box.click()

# #     input_box.fill(text)

# #     send_button = page.locator('button[data-testid="send-button"]')

# #     send_button.click()

# import pyperclip

# def send_prompt(page, text):

#     input_box = page.locator("#prompt-textarea")

#     input_box.wait_for(state="visible")

#     input_box.click()

#     pyperclip.copy(text)

#     page.keyboard.press("Control+V")

#     time.sleep(1)

#     page.keyboard.press("Enter")
# # ----------------------------
# # 获取 AI 回复
# # ----------------------------
# def get_last_answer(page):

#     messages = page.locator(
#         'div[data-message-author-role="assistant"]'
#     )

#     count = messages.count()

#     if count == 0:
#         return ""

#     return messages.nth(count - 1).inner_text()


# # ----------------------------
# # 等待 AI 输出结束
# # ----------------------------
# def wait_for_answer_finish(page, timeout=180):

#     last = ""

#     start = time.time()

#     while True:

#         text = get_last_answer(page)

#         if text == "":
#             time.sleep(2)
#             continue

#         if text == last:
#             return text

#         last = text

#         if time.time() - start > timeout:
#             return text

#         time.sleep(3)


# # ----------------------------
# # 保存回答
# # ----------------------------
# def save_answer(i, prompt, answer):

#     with open(RESULT_FILE, "a", encoding="utf-8") as f:

#         f.write("\n\n====================\n")
#         f.write(f"PROMPT {i+1}\n\n")
#         f.write(prompt)
#         f.write("\n\n---ANSWER---\n\n")
#         f.write(answer)


# # ----------------------------
# # 登录保存
# # ----------------------------
# def save_login(context):

#     print("请手动登录 ChatGPT")

#     input("登录完成按回车")

#     context.storage_state(path=AUTH_FILE)


# # ----------------------------
# # 主程序
# # ----------------------------
# def main():

#     prompts = extract_prompts("Udemy-UltimateAWSCertifiedSolutionsArchitectProfessional2025part2.md")

#     print("发现 prompt:", len(prompts))

#     start_index = load_progress()

#     print("从第", start_index + 1, "个开始")

#     with sync_playwright() as p:

#         browser = p.chromium.launch(
#             headless=False,
#             args=["--disable-blink-features=AutomationControlled"]
#         )

#         if os.path.exists(AUTH_FILE):

#             context = browser.new_context(
#                 storage_state=AUTH_FILE
#             )

#         else:

#             context = browser.new_context()

#         page = context.new_page()

#         page.goto(CHATGPT_URL)

#         wait_for_input(page)

#         if not os.path.exists(AUTH_FILE):
#             save_login(context)

#         for i in range(start_index, len(prompts)):

#             prompt = prompts[i]

#             print("\n===================")
#             print("PROMPT", i + 1)
#             print("===================")

#             print(prompt[:200])

#             send_prompt(page, prompt)

#             print("等待 AI 回复...")

#             answer = wait_for_answer_finish(page)

#             print(answer[:500])

#             save_answer(i, prompt, answer)

#             save_progress(i + 1)

#             sleep_time = random.randint(10, 20)

#             print("休息", sleep_time, "秒")

#             time.sleep(sleep_time)

#         print("全部完成")

#         browser.close()


# if __name__ == "__main__":
#     main()
from playwright.sync_api import sync_playwright
import time
import random
import os
import re
import pyperclip

CHATGPT_URL = "https://chat.openai.com"
RESULT_FILE = "result.txt"
PROGRESS_FILE = "progress.txt"

# ----------------------------
# 多账号配置
# ----------------------------
ACCOUNTS = {
    "1": "account1",  # 已登录账号
    "2": "account2",  # 新账号
}

# ----------------------------
# 提取 markdown prompt
# ----------------------------
def extract_prompts(md_file):
    with open(md_file, "r", encoding="utf-8") as f:
        content = f.read()
    pattern = r"```(.*?)```"
    prompts = re.findall(pattern, content, re.S)
    return [p.strip() for p in prompts if p.strip()]

# ----------------------------
# 读取/保存进度
# ----------------------------
def load_progress():
    if os.path.exists(PROGRESS_FILE):
        with open(PROGRESS_FILE) as f:
            return int(f.read().strip())
    return 0

def save_progress(i):
    with open(PROGRESS_FILE, "w") as f:
        f.write(str(i))

# ----------------------------
# 等待输入框
# ----------------------------
def wait_for_input(page):
    page.wait_for_selector('div[contenteditable="true"]', timeout=120000)

# ----------------------------
# 发送 prompt
# ----------------------------
def send_prompt(page, text):
    input_box = page.locator("#prompt-textarea")
    input_box.wait_for(state="visible")
    input_box.click()
    pyperclip.copy(text)
    page.keyboard.press("Control+V")
    time.sleep(1)
    page.keyboard.press("Enter")

# ----------------------------
# 获取 AI 回复
# ----------------------------
def get_last_answer(page):
    messages = page.locator('div[data-message-author-role="assistant"]')
    count = messages.count()
    if count == 0:
        return ""
    return messages.nth(count - 1).inner_text()

# ----------------------------
# 检测 AI 回复完成
# ----------------------------
def wait_for_answer_finish(page, timeout=180):
    last = ""
    start = time.time()
    while True:
        # 检查免费额度到期提示
        if page.locator("text=您已达到免费额度").count() > 0:
            return "FREE_QUOTA_EXCEEDED"

        text = get_last_answer(page)
        if text == "":
            time.sleep(2)
            continue
        if text == last:
            return text
        last = text
        if time.time() - start > timeout:
            return text
        time.sleep(3)

# ----------------------------
# 保存回答
# ----------------------------
def save_answer(i, prompt, answer):
    with open(RESULT_FILE, "a", encoding="utf-8") as f:
        f.write("\n\n====================\n")
        f.write(f"PROMPT {i+1}\n\n")
        f.write(prompt)
        f.write("\n\n---ANSWER---\n\n")
        f.write(answer)
    pyperclip.copy(answer)

# ----------------------------
# 登录保存
# ----------------------------
def save_login(context, storage_file):
    print("请手动登录 ChatGPT")
    input("登录完成后按回车继续...")
    context.storage_state(path=storage_file)
    print(f"登录状态已保存到 {storage_file}")

# ----------------------------
# 运行主程序
# ----------------------------
def main():
    prompts = extract_prompts("Udemy-UltimateAWSCertifiedSolutionsArchitectProfessional2025part3.md")
    print("发现 prompt:", len(prompts))

    start_index = load_progress()
    print("从第", start_index + 1, "个开始")

    # 初始化账号循环
    account_keys = list(ACCOUNTS.keys())
    current_account_index = 0

    with sync_playwright() as p:
        while start_index < len(prompts):
            account_key = account_keys[current_account_index]
            USER_DATA_DIR = ACCOUNTS[account_key]
            os.makedirs(USER_DATA_DIR, exist_ok=True)
            STORAGE_FILE = os.path.join(USER_DATA_DIR, "storage_state.json")

            print(f"\n使用账号 {account_key} ({USER_DATA_DIR}) 登录...")

            browser = p.chromium.launch(headless=False, args=["--disable-blink-features=AutomationControlled"])
            if os.path.exists(STORAGE_FILE):
                context = browser.new_context(storage_state=STORAGE_FILE)
            else:
                context = browser.new_context()

            page = context.new_page()
            page.add_init_script("""
                Object.defineProperty(navigator, 'webdriver', {
                    get: () => undefined
                })
            """)
            page.goto(CHATGPT_URL)
            wait_for_input(page)

            if not os.path.exists(STORAGE_FILE):
                save_login(context, STORAGE_FILE)

            # 循环执行 prompts
            for i in range(start_index, len(prompts)):
                prompt = prompts[i]
                print("\n===================")
                print("PROMPT", i + 1)
                print("===================")
                print(prompt[:200])

                send_prompt(page, prompt)
                print("等待 AI 回复...")
                answer = wait_for_answer_finish(page)

                # 检测到免费额度耗尽
                if answer == "FREE_QUOTA_EXCEEDED":
                    print(f"账号 {account_key} 免费额度已用完，切换账号...")
                    browser.close()
                    current_account_index += 1
                    if current_account_index >= len(account_keys):
                        print("所有账号额度已用完，退出。")
                        return
                    # 不更新 start_index，下一账号继续从同一个索引开始
                    break

                print(answer[:500])
                save_answer(i, prompt, answer)
                save_progress(i + 1)

                sleep_time = random.randint(10, 20)
                print("休息", sleep_time, "秒")
                time.sleep(sleep_time)

            else:
                # 当前账号完成所有 prompts
                browser.close()
                break

if __name__ == "__main__":
    main()
````
